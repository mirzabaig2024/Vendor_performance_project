```python
# Summarizing Sales by each Vendor and Brand
query = 'with freight_summary as(select vendornumber,sum(freight) as freigthcost from vendor_invoice group by vendornumber), purchase_summary as(select p.brand,p.vendorNumber,p.vendorname,p.description,p.purchaseprice,pp.price as actualprice,pp.volume,sum(p.quantity) As totalpurchasequantity,sum(p.dollars) as totalpurchasedollar from purchases p join purchase_prices pp on p.brand = pp.brand where p.purchaseprice > 0 group by p.vendornumber,p.vendorname,p.brand ),' \
' sales_summary as(select vendorno,brand,sum(salesdollars) as totalsaledollar, sum(salesprice) as totalpurchase,sum(salesquantity) as totalsalesquantity , sum(excisetax) as totalecisetax from sales  group by vendorno , brand ) ' \
'select ps.vendornumber,ps.vendorname,ps.brand,ps.description,ps.purchaseprice,ps.actualprice,ps.volume,ps.totalpurchasedollar ,ps.totalpurchasequantity,ss.totalsalesquantity,ss.totalsaledollar,ss.totalecisetax,fs.freigthcost from purchase_summary ps left join sales_summary ss on ps.vendornumber = ss.vendorno and ps.brand = ss.brand left join freight_summary fs on ps.vendornumber = fs.vendornumber order by totalpurchasedollar asc; '
final_summary = pd.read_sql(query,engine)
final_summary
```

# **Exploratory Data Analysis (EDA) of Final_Summary Table**
    Analyzing and Cleaning the Final_Summary Table

```python
# finding the null values in the final_summary table
final_summary.isnull().sum() #their are 178 null values in the final_summary table'''
```

```python
final_summary ['volume'] = final_summary ['volume'].astype('float') # initially volume datatype is objection with this query we change it into float
```

```python
#finding the unique vendor names in the final_summary table
final_summary ['vendorname'].unique()
```

```python
final_summary ['vendorname'] = final_summary ['vendorname'].str.strip() #initially there some space b/t them with using this query we removed the space b/t them
```

```python
final_summary.fillna(0,inplace= True) #replacing missing value 178 with 0
```

```python
final_summary.columns
```

```python
final_summary
```

```python
# Calculating Gross Profit and adding new column Gross_Profit
final_summary ['Gross_Profit'] = final_summary['totalsaledollar'] - final_summary['totalpurchasedollar']
```

```python
# calculating Profit Margin and adding new column ProfitMargin
final_summary ['ProfitMargin'] = (final_summary['Gross_Profit']/final_summary['totalsaledollar'])*100
```

```python
# calculating Stock Turnover and adding new column StockTurnover
final_summary ['stockturnover'] = final_summary['totalsalesquantity']/final_summary['totalpurchasequantity']
```

```python
# calculating Sales to Purchase Ratio and adding new column salestopurchaseratio
final_summary ['salestopurchaseratio'] = final_summary ['totalsaledollar']/final_summary['totalpurchasedollar']
```

```python
# replacing inf and -inf values with a large number to  import in mysql database
final_summary.replace([np.inf,-np.inf],1e300,inplace=True)
```

```python
# importing final_summary table into mysql database
final_summary.to_sql('final_summary',con=engine,if_exists='replace',index=False)
```
